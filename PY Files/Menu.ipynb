```python
button = widgets.Button(description="Click Me!")
def on_button_clicked():
          pass
button.on_click(on_button_clicked)

def searchForGame():


            digitalGameSearch = gs.loadGameSearch(["board", "digital"])
            options = []
            for game in digitalGameSearch.values():
                      options.append(game['title'])
                      options.sort()

            game_text = widgets.Text(
                      value='',
                      placeholder='Enter a game',
                      description='String:',
                      disabled=False,
                      continuous_update=False
            )

            bg_checkbox = widgets.Checkbox(
                      value=True,
                      description='Include board games',
                      disabled=False,
                      indent=False
            )

            dg_checkbox = widgets.Checkbox(
                      value=True,
                      description='Include digital games',
                      disabled=False,
                      indent=False
            )

            games_dropdown = widgets.Dropdown(
                      options=options,
                      value=options[0],
                      description='Select a game:',
                      disabled=False,
            )
            return [game_text, games_dropdown, bg_checkbox, dg_checkbox]
game_widgets = searchForGame()
widgets.HBox(game_widgets)
game_text.observe(searchForGame())

```
