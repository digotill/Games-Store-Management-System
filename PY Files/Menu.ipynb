```python
game_text = widgets.Text(value='', placeholder='e.g. minecraft', description='Enter a game:', disabled=False,continuous_update=True)
bg_checkbox = widgets.Checkbox(value=True, description='Include board games', disabled=False, indent=False)
dg_checkbox = widgets.Checkbox(value=True, description='Include digital games', disabled=False, indent=False)
game_radiobuttons = widgets.RadioButtons(options=[], description='Choose a game', disabled=False)
game_radiobuttons = widgets.ToggleButtons(options=[],description='Choose a game:',disabled=False,button_style='', tooltips=['Description of slow', 'Description of regular', 'Description of fast'],)

def on_search(change):
                    digitalGameSearch = gs.loadGameSearch(["board" if bg_checkbox.value else None, "digital" if dg_checkbox.value else None])
                    options = []
                    for game in digitalGameSearch.values():
                                        if game_text.value.lower() in game['title'].lower() or game_text.value == "":
                                                  options.append(game['title'])

                    game_radiobuttons.options = options

on_search("")
bg_checkbox.observe(on_search, names='value')
dg_checkbox.observe(on_search, names='value')
game_text.observe(on_search, names='value')


widgets_container1 = HBox([game_text, bg_checkbox, dg_checkbox], layout=Layout(display='flex'))
widgets_container2 = HBox([game_radiobuttons], layout=Layout(display='flex'))
widgets_container3 = HBox([], layout=Layout(display='flex'))

tab_content_1 = VBox([widgets_container1, widgets_container2])


tab_widget = widgets.Tab()
tab_widget.children = [tab_content_1, widgets.Label("Content for Tab 2"), widgets.Label("Content for Tab 3")]
tab_widget.titles = ["Search", "Renting", "Booking"]
display(tab_widget)

```
