5 Code Security Issues

1. Input Validation and Data Sanitization (Injection Risks)
        This makes my system vulnerable, even though im not using SQL, they can input python code snyppets or shell commands to alter the functions of my system. I can Prevent this by doing lots of input validation.
2. Weak Customer ID and Subscription Checks
        Since the Customer ID is only a 4 digit string that only contains letters it would be easy for a hacker to brute force access to the system, allowing access to personal information. I can prevent this by using stronger ID system like UUID making usernames more difficult to guess.
3. Exposure of Sensitive/Personal Data
        Since my databases are stored in just text files, they would be very easy to view and alter if a hacker were to gain access to them. Encrypting them is a safe way to prevent this.
4. Inventory Pruning Logic Flaws (Denial of Service/Logic Bomb)
        If the criteria for removing games is flawed it can lead to DoS but deleting too many essential games, an attacker could effectivly estroy the system if they gained access to the Pruning feature.
5. Exposure of Sensitive/Personal Data
        Since all of the files in this project are pure .py files, they are completly readable and alterable, meaning a hacker could easily view and change them if they wish. To avoid this i could compile the files so the funcionality of modules cannot be changed. Additionally, i could attempt to hide the files by putting them inside other files no one would think to look in.
