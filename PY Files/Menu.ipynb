```python
game_text = widgets.Text(value='', placeholder='Enter a game', description='String:', disabled=False,continuous_update=False)
bg_checkbox = widgets.Checkbox(value=True, description='Include board games', disabled=False, indent=False)
dg_checkbox = widgets.Checkbox(value=True, description='Include digital games', disabled=False, indent=False)
search_button = widgets.Button(description='Search', disabled=False, button_style='', tooltip='Click me', icon='check')

game_radiobuttons = widgets.RadioButtons(options=[], description='Choose a game', disabled=False)

widgets_container1 = HBox([game_text, search_button, bg_checkbox, dg_checkbox], layout=Layout(display='flex'))
widgets_container2 = VBox([game_radiobuttons], layout=Layout(display='none'))


def on_button_click(b):
                    widgets_container2.layout.display = 'flex'
                    b.button_style = 'success'

                    digitalGameSearch = gs.loadGameSearch(["board" if bg_checkbox.value else None, "digital" if dg_checkbox.value else None])
                    options = []
                    for game in digitalGameSearch.values():
                                    if game_text.value in game['title'].lower():
                                                  options.append(game['title'])
                    options.sort()

                    game_radiobuttons.options = options

search_button.on_click(on_button_click)

display(widgets_container1, widgets_container2)

```

```python
import ipywidgets as widgets
from ipywidgets import Button, VBox, IntSlider, Label, Layout, HTML
from IPython.display import display

# --- 1. Target Widgets and Container (Same as before) ---
slider = IntSlider(description='Value:', min=0, max=10, step=1, value=5)
label = Label(value="These widgets are now visible!")
text_area = widgets.Textarea(value='Example text area.', rows=3)

# Container for the content to be shown/hidden
widgets_container = VBox(
    [label, slider, text_area],
    layout=Layout(display='none')
)

# Control Button
toggle_button = Button(
    description='Show Widgets',
    button_style='info',
)

# Callback Function
def on_button_click(b):
    if widgets_container.layout.display == 'none':
        widgets_container.layout.display = 'flex'
        b.description = 'Hide Widgets'
        b.button_style = 'success'
    else:
        widgets_container.layout.display = 'none'
        b.description = 'Show Widgets'
        b.button_style = 'info'

# Link the Button to the Function
toggle_button.on_click(on_button_click)

# --- 2. Create the Content for the Tab ---
# We use a VBox to vertically arrange the introductory HTML, the button, and the hidden container.
# This entire VBox will become the content of our first tab.
tab_content_1 = VBox([
    HTML("<h4>Click the button to manage the hidden controls below:</h4>"),
    toggle_button,
    widgets_container
])

# --- 3. Create the Tab Container ---
tab_widget = widgets.Tab()

# Assign the content VBox to the Tab's 'children' property.
# You can add more VBox containers for additional tabs here.
tab_widget.children = [tab_content_1, widgets.Label("Content for Tab 2"), widgets.Label("Content for Tab 3")]

# --- 4. Set the Titles for the Tabs ---
# The number of titles must match the number of children.
tab_widget.set_title(0, 'Toggle Interface')
tab_widget.set_title(1, 'Other Settings')
tab_widget.set_title(2, 'Data View')

# --- 5. Display the Tab Widget ---
display(tab_widget)
```
