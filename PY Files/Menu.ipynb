```python
import ipywidgets as widgets
from IPython.display import display
from ipywidgets import VBox, Layout, HBox

import feedbackManager as fm
import gameSearch as gs
import subscriptionManager as sm
```

```python
game_text = widgets.Text(value='', placeholder='e.g. minecraft', description='Enter a game:', disabled=False,continuous_update=False)
bg_checkbox = widgets.Checkbox(value=True, description='Include board games', disabled=False, indent=False)
dg_checkbox = widgets.Checkbox(value=True, description='Include digital games', disabled=False, indent=False)
search_button = widgets.Button(description='Search', disabled=False, button_style='', tooltip='Click me', icon='check')

game_radiobuttons = widgets.RadioButtons(options=[], description='Choose a game', disabled=False)
choose_button = widgets.Button(description='Choose', disabled=False, button_style='', tooltip='Click me', icon='check')
html1 = widgets.HTML(value="No games found",placeholder='',description=' ',)

widgets_container1 = HBox([game_text, search_button, bg_checkbox, dg_checkbox], layout=Layout(display='flex'))
widgets_container2 = HBox([game_radiobuttons, choose_button], layout=Layout(display='none'))
widgets_container3 = HBox([html1], layout=Layout(display='none'))


def on_search_button_click(b):
                    if widgets_container2.layout.display == 'none':
                                    widgets_container2.layout.display = 'flex'
                                    b.description = 'Cancle'
                                    b.button_style = 'success'

                                    digitalGameSearch = gs.loadGameSearch(["board" if bg_checkbox.value else None, "digital" if dg_checkbox.value else None])
                                    options = []
                                    for game in digitalGameSearch.values():
                                                    if game_text.value.lower() in game['title'].lower():
                                                                  options.append(game['title'])
                                    options.sort()
                                    if len(options) == 0:
                                              widgets_container3.layout.display = 'flex'

                                    game_radiobuttons.options = options
                    else:
                                    widgets_container2.layout.display = 'none'
                                    widgets_container3.layout.display = 'none'
                                    b.description = 'Search'
                                    b.button_style = 'info'

search_button.on_click(on_search_button_click)

tab_content_1 = VBox([widgets_container1, widgets_container2])

tab_widget = widgets.Tab()

tab_widget.children = [tab_content_1, widgets.Label("Content for Tab 2"), widgets.Label("Content for Tab 3")]

tab_widget.titles = ["Search Tool", "2", "3"]

display(tab_widget)

```
